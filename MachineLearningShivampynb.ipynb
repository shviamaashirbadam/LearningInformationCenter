```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_california_housing
import seaborn as sns
# import matplotlib.pyplot as plt
from matplotlib import pyplot as plt 
californiaData = fetch_california_housing(as_frame=True)
X, y = fetch_california_housing(return_X_y=True)
X = X[:2000,:]
y = y[:2000]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# print(X_test.shape)
# print(y_test.shape)
# print(californiaData.frame.head(10))
# print(californiaData.frame.info)
# californiaData.frame.info

# X_train.d



# plt.plot(X, y, "b.")
# plt.xlabel("$x$", fontsize=18)
# plt.ylabel("$y$", fontsize=18)
# plt.tight_layout()
# plt.show()


lreg = LinearRegression()
lr = lreg.fit(X_train, y_train)
y_pred = lreg.predict(X_test)

plt.scatter(y_test,y_pred);
plt.xlabel('Actual');
plt.ylabel('Predicted');
sns.regplot(x=y_test,y=y_pred,ci=None,color ='green');
# plt.scatter(X_train, y_train,color='g') 



# plt.plot(X_test, y_pred,color='g')
# plt.show()
# print("Accuracy Score:", lr.score(X_test, y_test)*100)

# plt.scatter(X['total_bedrooms'], y)
# plt.xlabel('total_bedrooms')
# plt.ylabel('House Price')


```
